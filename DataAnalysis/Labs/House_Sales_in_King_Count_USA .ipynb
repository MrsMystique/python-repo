# Module 4: Model Development


Мы можем установить модель линейной регрессии, используя функцию долготы <code> 'long' </code> и Caculate r^2.


```python
X = df[['long']]
Y = df['price']
lm = LinearRegression()
lm.fit(X,Y)
lm.score(X, Y)
```

Вызывается метод lm.fit(X,Y), который обучает линейную регрессию на основе X и Y и находит оптимальные параметры модели, которые максимизируют функцию потерь.

Наконец, вызывается метод lm.score(X, Y), который вычисляет коэффициент детерминации (R^2 score) для обученной модели. Коэффициент детерминации показывает, насколько хорошо модель соответствует данным: значения от 0 до 1, где 1 означает идеальное соответствие.
