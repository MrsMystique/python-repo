### индексация - ILOC и LOC

Очень важно понять различия между LOC и ILOC.
Глядя на следующую ячейку, можно подумать, что они делают то же самое.

(Когда вы запросите только на одну строку, вы получаете объект, чье имя является индексом выбранной строки.)

````python
df.iloc[10] == df.loc[10]
````

````python
all(df.iloc[10] == df.loc[10])
````

В первый раз, когда мы загрузили CSV в DataFrame, мы не сказали Pandas интерпретировать конкретный столбец как индекс, поэтому Pandas создал для нас индекс.Следует ли уважать упорядочение, наложенное на наши данные этим индексом, является вопросом, по которому «ILOC» и «LOC» не согласованы.

** Чтобы действительно узнать разницу `iloc` и` loc`, нам нужно перетасовать ряды **


В данном коде используется метод `sample` для перестановки случайного порядка строк в DataFrame `df`. Ниже приведено объяснение параметров:

- `frac`: Определяет долю строк, которые нужно выбрать. В данном случае, `frac=1` означает, что нужно выбрать все строки DataFrame.

- `random_state`: Определяет начальное значение для генератора случайных чисел. Это позволяет сохранить одинаковые случайные порядки строк каждый раз, когда код выполняется с одним и тем же значением `random_state`. В данном случае, `random_state=109` используется для получения одного и того же случайного порядка строк при каждом запуске кода.

Таким образом, данный код переставляет строки DataFrame `df` в случайном порядке, используя генератор случайных чисел с начальным значением 109.

````python
df = df.sample(frac=1, random_state=109)
df
````

Теперь давайте повторим наш код из ранее.

````python
all(df.iloc[10] == df.loc[10])
````

Оказывается, что `loc` - фильтры по значению индекса (что - то вроде` df.index == value`).

То есть, результаты Loc` зависят только от **индексов**. Фактические позиции элементов в DataFrame игнорируются.

````python
df.loc[10]
````

Напротив, `iloc` - фильтры по положению строки (что -то вроде` df.index == df.index [value] `)

Таким образом, результаты `iloc` зависят от **фактических позиций** элементов в структуре данных Pandas.
Индексы этих элементов игнорируются.

Именно эта разница объясняет противоречивые результаты, подобные этим:

````python
df.index[10]
````

Рассмотрим индекс одной строки с `iloc`

````python
df.iloc[10]
````

### Категориальные столбцы - nunique () и уникальный ()

Если некоторые из ваших данных являются категориальными (например, принятие набора значений описания с присущим заказом), вы часто узнаете, сколько уникальных значений вы имеете дело.`nunique` и` уникальные 'здесь, чтобы помочь и могут использоваться в одном столбце или в нескольких столбцах.

**Exercise**  
Заполните пробелы, используя `unique ()` и `nunique ()` для завершения метода `print_uniques`, чтобы помочь нам узнать больше о категориальных переменных в наших данных.

````python
### edTest(test_f) ###
def print_uniques(df, col, limit=10):
    """Print column's uniques values when the number of column's uniques is lower or equal than limit """
    n = df[col].nunique()
    if n <= limit:
        print(f'{col}:', df[col].unique().tolist())
    else:
        print(f'{col}:', f'more than {limit} uniques')

for col in df.columns:
    print_uniques(df, col)
````
