Beautiful Soup - это библиотека Python, которая используется для извлечения данных из HTML и XML файлов. Она предоставляет простой способ навигации, поиска и модификации древовидной структуры документа.

Вот некоторые часто используемые методы, их описание и примеры кода для парсинга данных с помощью Beautiful Soup:

1. BeautifulSoup - инициализирует объект Beautiful Soup и загружает в него HTML или XML файл.
   Пример: 
   ```python
   from bs4 import BeautifulSoup
   with open("example.html") as file:
       soup = BeautifulSoup(file, 'html.parser')
   ```

2. find - ищет первый элемент, соответствующий указанным критериям.
   Пример:
   ```python
   result = soup.find('div', class_='content')
   ```

3. find_all - находит все элементы, соответствующие указанным критериям, и возвращает их в виде списка.
   Пример:
   ```python
   results = soup.find_all('a')
   ```

4. select - выполняет выборку элементов с помощью CSS-селекторов.
   Пример:
   ```python
   results = soup.select('.content')
   ```

5. parent - возвращает родительский элемент текущего элемента.
   Пример:
   ```python
   parent = result.parent
   ```

6. contents - возвращает список всех прямых потомков элемента.
   Пример:
   ```python
   children = result.contents
   ```

7. text - извлекает текстовое содержимое элемента.
   Пример:
   ```python
   text_content = result.text
   ```

8. get - извлекает значение атрибута элемента.
   Пример:
   ```python
   attr_value = result.get('href')
   ```

9. next_sibling - возвращает следующий элемент на том же уровне вложенности.
   Пример:
   ```python
   next_element = result.next_sibling
   ```

10. previous_sibling - возвращает предыдущий элемент на том же уровне вложенности.
    Пример:
    ```python
    prev_element = result.previous_sibling
    ```

Это лишь некоторые из множества методов, предлагаемых Beautiful Soup. Работая с этой библиотекой, можно получить доступ к различным атрибутам и методам элементов, что позволяет более эффективно парсить и анализировать HTML и XML данные. Более подробную информацию о методах можно найти в официальной документации Beautiful Soup.

````python
from bs4 import BeautifulSoup
import requests

# Получить HTML-страницу
url = 'https://apnews.com/hub/ap-top-news'
response = requests.get(url)
html = response.content

# Создать объект BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Найти все уникальные классы
unique_classes = set()
for tag in soup.find_all():
    if tag.has_attr('class'):
        for cls in tag['class']:
            unique_classes.add(cls)

# Найти все уникальные теги
unique_tags = set()
for tag in soup.find_all():
    unique_tags.add(tag.name)

print("Уникальные классы:")
for cls in unique_classes:
    print(cls)

print("\nУникальные теги:")
for tag in unique_tags:
    print(tag)
````

9. Извлечение данных из RSS-ленты:


````python

rss_feed = 'https://www.example.com/rss-feed.xml'
response = requests.get(rss_feed)
rss = response.text

soup = BeautifulSoup(rss, 'xml')
items = soup.find_all('item')
for item in items:
    title = item.find('title').text
    print(title)
````

10. Извлечение данных из HTML-таблицы в формате CSV:


````python
import csv

table = soup.find('table')
rows = table.find_all('tr')

with open('output.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for row in rows:
        columns = row.find_all('td')
        row_data = [column.text for column in columns]
        writer.writerow(row_data)
````
