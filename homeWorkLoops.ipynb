**Задача 10:** На столе лежат n монеток. Некоторые из них лежат вверх решкой, а некоторые – гербом. 

Определите минимальное число монеток, которые нужно перевернуть, чтобы все монетки были повернуты вверх одной и той же стороной. 

Выведите минимальное количество монет, которые нужно перевернуть

```python
# импортну рандом
import random
# принять количество монет с консоли. Часть лежит решкой. часть гербом вверх (судя по всему в ход идет рандом)
try:
    number_of_coins = int(input("Введите количество монет: "))
    # создам пустой список для добавления разныхх сторон монет
    coins_list = []
    # циклом буду рандомно выбрасывать монетку и класть значение (сторону) в список
    for coin in range(number_of_coins):
        coin_side = coins_list.append(random.choice(["Орел", "Решка"]))
    print(f"Наши {number_of_coins} монет лежат следующим образом {coins_list}")
# посчитаю значения с помощью count()(сколько монеток лежат той или иной стороной)  
# функцией min() выберу минимальное. это будет ответом на вопрос сколько минимум нужно перевернуть. 
    print(f"Минимальное количество монет, которое необходимо перевернуть для того чтобы они лежали одной стороной: {min(coins_list.count('Орел'), coins_list.count('Решка'))}")
except ValueError:
    print("Что-то пошло не так. Это не целое число")
```

**Задача 12:** Петя и Катя – брат и сестра. Петя – студент, а Катя – школьница. Петя помогает Кате по математике. 

Он задумывает два натуральных числа X и Y (X,Y≤1000), а Катя должна их отгадать. Для этого Петя делает две подсказки. Он называет сумму этих чисел S и их произведение P. 

Помогите Кате отгадать задуманные Петей числа.

```python
# Получаем от Пети сумму и произведение двух чисел, каждое из которых меньше либо равное 1000 
Sum = int(input("Введите сумму двух чисел: "))
Product = int(input("Введите произведение двух чисел: "))
# Раз уж задано ограничение в условии..предположим, что сумма не должна превышать 2000
while Sum > 2000:
    Sum = int(input("Введите сумму двух чисел не превышающую 2000: "))
    Product = int(input("Введите произведение двух чисел (до млн): "))
# Перебираем все пары чисел (и одно и второе) в диапазоне суммы первое число, произведения - второе число
for number in range(Sum):
    for number2 in range(Product):
        # Если сумма и произведение пары чисел равны заданным значениям, то выводим результат и завершаем программу
        if number + number2 == Sum and number * number2 == Product:
            print("Первое число, которое загадал Петя:", number)
            print("Второе число, которое загадал Петя:", number2)
            
```

**Задача 14:** Требуется вывести все целые степени двойки (т.е. числа вида 2k), не превосходящие числа N.

```python
# примем число от пользованиеля - ограничение результата возведения в степень двойки
NumberMax = int(input("Введите максимальный результат возведения в степень двойки: "))
# текущая (начальная) степень - 0
current_degree = 0
# Используем оператор возведения в степень для проверки истинности условия по ограничению
while 2 ** current_degree <= NumberMax:
    ## Выводим результат возведения в каждую степень
    print(f"2 в степени {current_degree} = {2**current_degree}")
    # инкриментим степень пока while не вернет false
    current_degree += 1
print(f"Следующее возведение в степень превысит максимум, равный {NumberMax}")
```
