# Теория вероятностей и математическая статистика 
**Урок 2. Дискретные распределения вероятностей**

Основная формула:

Используя формулу биномиального распределения, вероятность можно вычислить следующим образом:

P(X = k) = C(n, k) * p^k * (1 - p)^(n - k)

где P(X = k) - вероятность положительных исходов(удовлетворяющих нас событий),
C(n, k) - количество сочетаний из n по k, или биномиальный коэффициент,
p - вероятность возникновения события в одном испытании,
n - общее количество событий.

1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

p = 0.8 - вероятность возникновения положительного исхода( вероятность попадания в цель)

n = 100 - общее количество выстрелов(испытаний)

k = 85 - положительные исходы (попадания)

$$profability = C_{100}^{85}\cdot{0.8}^{85}\cdot{0.2}^{15} =0.048 $$



Из 100 выстрелов, 85 являются попаданиями и 15 - промахами. 

Мы хотим найти вероятность того, что из 100 выстрелов мы делаем 85 попаданий и 15 промахов. 

Для этого мы будем использовать формулу для числа сочетаний (C(n, k)), где n - общее количество выстрелов, а k - количество пападаний.

В данном случае, мы хотим выбрать 85 попаданий из 100, поэтому мы вычисляем C(100, 85), используя формулу для сочетаний.

$$C(100, 85) = \frac{100!}{85!(100-85)!}$$

Также, нам дано, что вероятность попасть при выстреле равна 0.8, следовательно, вероятность промахнуться равна 0.2.

Таким образом, вероятность совершить 85 попаданий из 100 выстрелов равна:
$$P = C(100, 85) \cdot (0.8^{85}) \cdot (0.2^{15})$$

Подставляя значения в формулу, получаем:

$$P = \frac{100!}{85!(100-85)!}\cdot(0.8^{85})\cdot(0.2^{15})$$



```python
import math

k = 85
n = 100
p = 0.8

probability = math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
print(f"Вероятность попасть ровно 85 раз: {probability}")

percentage = probability * 100

print(f"Процентная веротяность: {percentage:.2f}%")
```

2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. 
В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? 
Какова вероятность, что перегорят ровно две?

$$P(X = 0) = (1 - p)^n$$

```python
p = 0.0004
n = 5000

# C
probability_no_gone = (1 - p) ** n
print(f"процентная верноятность что ни одна лампочка не перегорит{probability_no_gone}, Процентная веротяность: {probability_no_gone*100:.2f}%")

# Вероятность, что перегорят ровно две лампочки
k = 2
probability_two_gone = math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

print(f"Процентная веротяность что перегорят 2 лампочки: {probability_two_gone*100:.2f}%")
```

3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?


```python
k = 70
n = 144
p = 0.5

probability = math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
print(f"Вероятность того, что орел выпадет ровно 70 раз: {probability}")
print(f"Процентная веротяность: {probability*100:.2f}%")
```

4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. 

Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? 

Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

```python
# а) Вероятность того, что все мячи будут белыми
favorable_outcomes_a = math.comb(7, 2) * math.comb(9, 2)
total_outcomes_a = math.comb(10, 2) * math.comb(11, 2)
probability_a = favorable_outcomes_a / total_outcomes_a

# б) Вероятность того, что ровно два мяча будут белыми
favorable_outcomes_b = (math.comb(7, 2) * math.comb(3, 0) * math.comb(2, 2) * math.comb(9, 2)) + (math.comb(7, 0) * math.comb(3, 2) * math.comb(2, 2) * math.comb(9, 2))
total_outcomes_b = math.comb(10, 2) * math.comb(11, 2)
probability_b = favorable_outcomes_b / total_outcomes_b

# в) Вероятность того, что хотя бы один мяч будет белым
favorable_outcomes_c = math.comb(3, 2) * math.comb(2, 2)
total_outcomes_c = math.comb(10, 2) * math.comb(11, 2)
probability_c = 1 - (favorable_outcomes_c / total_outcomes_c)

print(f"Вероятность, что все мячи белые: {probability_a}")
print(f"Вероятность, что ровно два мяча белые: {probability_b}")
print(f"Вероятность, что хотя бы один мяч белый: {probability_c}")
```
