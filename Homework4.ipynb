Задача 24: В фермерском хозяйстве в Карелии выращивают чернику. Она растет на круглой грядке, причем кусты высажены только по окружности. Таким образом, у каждого куста есть ровно два соседних. Всего на грядке растет N кустов.

Эти кусты обладают разной урожайностью, поэтому ко времени сбора на них выросло различное число ягод – на i-ом кусте выросло ai ягод.

В этом фермерском хозяйстве внедрена система автоматического сбора черники. Эта система состоит из управляющего модуля и нескольких собирающих модулей. Собирающий модуль за один заход, находясь непосредственно перед некоторым кустом, собирает ягоды с этого куста и с двух соседних с ним.

Напишите программу для нахождения максимального числа ягод, которое может собрать за один заход собирающий модуль, находясь перед некоторым кустом заданной во входном файле грядки.

```python
# задача - найти макс сумму куста и 2 соседних
import random
print("Введите количество кустов от 3 до 1000: ")
numberShrubs = int(input("Введите количество кустов"))
CurSum = 0
MaxSum = 0
i = 0
while (numberShrubs < 3 or numberShrubs > 1000):
    print("\nКоличество кустов должно быть от 3 до 1000\nвведите верное значение: ")
    numberShrubs = int(input("Введите количество кустов"))

print (f"Количество кустов  {numberShrubs}")
CountBerrys = []
# Заполним список рандомно (сколько ягод на каждом кусте)
CountBerrys = [random.randint(1,100) for _ in range(numberShrubs)]
print(f"Вывожу количество ягод на каждом кусте: {CountBerrys}")
# нужно проитерироваться и складывать соседние элементы. типа индекс 0, значит -1+0+1 и определить макс с умму
for shrub in range(numberShrubs-1):
   CurSum = CountBerrys [shrub] + CountBerrys[shrub-1] + CountBerrys[shrub+1]
   if CurSum > MaxSum:
       MaxSum = CurSum
print(f"Максимальная сумма ягод, которые возможно собрать за один заход: {MaxSum}")
```
