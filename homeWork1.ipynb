**Задача 2:** Найдите сумму цифр трехзначного числа.

*Пример:*

123 -> 6 (1 + 2 + 3)

100 -> 1 (1 + 0 + 0) |

```python
# приму число в виде строки для проверки длины. 
number = input("Введите трехзначное число: ")

while len(number) != 3:
    print(f"Ошибка! {number} - это не корректные данные")
    number = input("Введите корректное число (трехзначное): ")
    
# проработаю ситуацию где длинна 3, но пользователь ввел символы
try:
    sum = int(number[0]) + int(number[1]) + int(number[2])
    print("Сумма цифр трехзначного числа", number, "равна", sum)
except ValueError:
    print(f"Что то пошло не так, возможно {number} - это не число")
```

**Задача 4:** Петя, Катя и Сережа делают из бумаги журавликов. Вместе они сделали S журавликов. Сколько журавликов сделал каждый ребенок, 

если известно, что Петя и Сережа сделали одинаковое количество журавликов, а Катя сделала в два раза больше журавликов, чем Петя и Сережа вместе?

*Пример:*

6 -> 1  4  1

24 -> 4  16  4

60 -> 10  40  10

```python
# запрошу общее количество журавликов
s = input("Введите число журавликов, которое сделали дети: ")
# s = 2*x + 2(2*x) = 6x
# все числа, которые делятся на 6 четные
# юзер может ввести символы
try:
    while int(s) < 6 or int(s) % 2!= 0:
        print("Введите четное число, которое больше или равное 6 (шести)")
        s = input("Введите четное число журавликов, которое равно или первышает 6: ")
    print(f"Мальчики сделали по {round(int(s)/6)} журавликов, а Катя сделала {round((int(s)/6)*4)}")
except ValueError:
    print(f"Что-то пошло не так. возможно {s} - это не число")
```

**Задача 6:** Вы пользуетесь общественным транспортом? Вероятно, вы расплачивались за проезд и получали билет с номером. 

Счастливым билетом называют такой билет с шестизначным номером, где сумма первых трех цифр равна сумме последних трех. 

Т.е. билет с номером 385916 – счастливый, т.к. 3+8+5=9+1+6. Вам требуется написать программу, которая проверяет счастливость билета.

*Пример:*

385916 -> yes

123456 -> no


```python
# приму в форме строки для того чтобы оценить длину функцией len()
ticket = input("Введите номер билета: ")
# что если введет не 6 знасчный. обработаем
while len(ticket) != 6:
    print(f"Ошибка. Вы ввели {ticket}, а это не номер билета")
    ticket = input("Введите корректный шестизначный номер: ")
# просуммируем элементы  (по 3 числа с каждой стороны) конвертируя их в инты
sum_left = int(ticket[0])+ int(ticket[1]) + int(ticket[2])
sum_right = int(ticket[3]) + int(ticket[4]) + int(ticket[5])
# проверка на здравый смысл (можно удалить)
print(f"Сумма чисел {ticket[:3]}  = {sum_left}") 
print(f"Сумма чисел {ticket[3:]}  = {sum_right}")
# ветвление с проверкой на равенство двух сумм
if (sum_left == sum_right):
    print(f"Билет номер {ticket} - счастливый")
else:
    print("Везение не на вашей стороне")
```

**Задача 8:** Требуется определить, можно ли от шоколадки размером n × m долек отломить k долек, если разрешается сделать один разлом 

по прямой между дольками (то есть разломить шоколадку на два прямоугольника).

*Пример:*

3 2 4 -> yes

3 2 1 -> no

```python
# функция рисует кусочки шоколадки
def print_square(m,n):
    # для каждой строки
    for i in range(n):
        # Для каждого столбца
        for j in range(m):
            #  рисовать кусочки
            print("#", end="")
        print()
              
n = int(input("Введите количество долек по оси х: "))
m = int(input("Введите количество долек по оси у: "))
k = int(input("Введите количество долек, которое нужно отломить: "))
# наша шоколадка
print_square(m,n)

# Проверяем, можно ли отломить k долек (количество долек должно делиться на сторону(либо одну, либо другую) без остатка 
# шоколадка должна быть не меньше чем кол-во долек)
if k <= n * m and (k % n == 0 or k % m == 0):
    print("Да, можно отломить", k, "долек")
else:
    print("Нет, нельзя отломить", k, "долек")
```
