1. Из колоды в 52 карты извлекаются случайным образом 4 карты. 

* Найти вероятность того, что все карты – крести. 
* Найти вероятность, что среди 4-х карт окажется хотя бы один туз.


52 карты (полная колода, от двоек до тузов) — используется для игры в покер; 

В такой колоде карт 13 карт масти крести

Чтобы найти вероятность того, что из 4 извлеченных карт все будут крести, нам нужно разделить количество благоприятных исходов на общее количество исходов.

Общая формула: 

<code> Вероятность события = (кол-во исходов, при которых событие произойдёт) / (общее кол-во исходов)</code>

<code> P(A) = (кол-во исходов, при которых происходит событие A) / (общее кол-во исходов)</code>

в нашем случае - будем считать:
* шанс выпадения 4 крест из 13 возможных ***(C(4, 13) - количество комбинаций из 13 по 4.)***
(13!/4!*(13!-4!)) = 715
* В данном случае, мы хотим найти количество комбинаций выбрать 4 карты из колоды в 52 карты **(C(4, 52) - количество комбинаций из 52 по 4)**
(52!/4!*(52!-4!)) = 270725
* шанс выпадения 4 карт крести из возможных вариантов

Теперь мы можем вычислить вероятность каждого события: C1(n1, m1)/C2(n2, m2) = 715/2070725 = 0.002641056422569
где n - количество удовлетворяющих исходов, а m - количество возможных исходов

$$C_{m}^{n}= \frac{{m!}}{{n!*(m!-n!)}}$$


```python
import math

favorable_outcomes = math.comb(13, 4)
total_outcomes = math.comb(52, 4)

probability = favorable_outcomes / total_outcomes

print(f"Вероятность того, что все карты - крести: {probability}")

percentage = probability * 100

print(f"Процентная веротяность: {percentage:.2f}%")
```

Когда мы говорим о "хотя бы одном тузе", это означает, что у нас может быть 1 туз, 2 туза, 3 туза или 4 туза. Поэтому благоприятными исходами будут все комбинации, где извлечены 1, 2, 3 или 4 туза.

Общее количество благоприятных исходов будет равно сумме следующих комбинаций:
* Комбинация 1: 1 туз из 4 (C(4, 1)) и 3 карт, которые не являются тузами (C(48, 3))
* Комбинация 2: 2 туза из 4 (C(4, 2)) и 2 карты, которые не являются тузами (C(48, 2))
* Комбинация 3: 3 туза из 4 (C(4, 3)) и 1 карта, которая не является тузом (C(48, 1))
* Комбинация 4: 4 туза из 4 (C(4, 4))

```python
favorable_outcomes = math.comb(4, 1) * math.comb(48, 3) + math.comb(4, 2) * math.comb(48, 2) + math.comb(4, 3) * math.comb(48, 1) + math.comb(4, 4)
total_outcomes = math.comb(52, 4)

probability = favorable_outcomes / total_outcomes

print(f"Вероятность, что среди 4 карт окажется хотя бы один туз: {probability}")
print(f"Процентная вероятность: {probability * 100:.2f}%")
```

Вариант второй: найти вероятность выпадения туза (С(4,52))
Р(А)= 52!/(4!*(52!-4!))
так как тянем 4 раза то вероятность увеличивается на 4, то есть Р(А)*4, но неточность будет состоять в том, что каждый раз при вытягивании НЕ ТУЗА, вероятность его выпадения увеличивается, так так как количяество вариантов уменьшается

2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?


```python
favorable_outcomes = 1 # 3 из 3
total_outcomes = math.comb(10, 3)

probability = favorable_outcomes / total_outcomes

print(f"Вероятность открыть дверь с первой попытки: {probability}")
print(f"Процентная веротяность: {probability * 100:.2f}%")
```

3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?
всего возможных исходов исходов 15
окрашенных - 9 
количество испытаний (благоприятных исходов) 3




```python
favorable_outcomes = math.comb(9, 3)
total_outcomes = math.comb(15, 3)

probability = favorable_outcomes / total_outcomes

print(f"Вероятность, что все извлеченные детали окрашены: {probability}")
print(f"Процентная веротяность: {probability * 100:.2f}%")
```

4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Количество благоприятных исходов (ситуаций, когда оба приобретенных билета выигрышные) будет равно 1, так как есть только одна комбинация, когда оба билета будут выигрышными.

P(оба приобретенных билета выигрышные) = (количество благоприятных исходов) / (общее количество исходов)


```python
favorable_outcomes = 1 # 2 из 2, то есть 2/2
total_outcomes = math.comb(100, 2)

probability = favorable_outcomes / total_outcomes

print(f"Вероятность, что 2 приобретенных билета окажутся выигрышными: {probability}")
print(f"Процентная веротяность: {probability * 100:.2f}%")
```
