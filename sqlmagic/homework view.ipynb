```python
%%sql
create table if not exists trains (
    train_id int not null,
    station_character varchar(50) not null,
    station_time_without_timezone time not null,
    time_to_next_station_interval time
)
```

```python
%%sql
alter table trains
add column time_to_next_station time
insert into train(time_to_next_station)
values()
```

```python
%%sql

SELECT
    train_id,
    station_character,
    station_time_without_timezone,
    SUBTIME(
        LEAD(station_time_without_timezone) OVER(PARTITION BY train_id ORDER BY train_id),
        station_time_without_timezone
    ) as time_to_next_station
FROM trains
```
